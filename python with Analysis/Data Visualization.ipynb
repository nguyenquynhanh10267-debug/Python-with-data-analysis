# 1 Making a Pivot Table

.pivot() Return reshaped DataFrame organized by given index/column values

```python
# dropping null values
df.dropna(inplace = True)
# ghi đè nội dung bên trong khung dữ liệu, tìm hiểu thêm
```

```python
# making a pivot table
df_pivot = df.pivot(
    index = 'year',
    columns = 'country',
    values = 'population'
)
```

```python
# selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 'Vietnam']]
```

```python
# showing pivot table
df_pivot
```

# 2 Lineplot

```python
df_pivot.plot(kind = 'line')
#  hiểu thêm chức năng plot()
```

```python
df_pivot.plot(kind = 'line', xlabel = 'Year', ylabel = 'Population',
              title = 'Population (1955-2020)', figsize = (8,4) )
```

# 3 Barplot

## 3.1 Single Barplot

```python
# selecting only one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020.0])]
```

```python
# showinh new dataframe
df_pivot_2020
```

```python
# transpose data frame from rowa to columns or viceversa
# phương pháp chuyển vị
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind='bar', color = 'brown', xlabel='country',
                    ylabel= 'population', title=' population 2020')
```
