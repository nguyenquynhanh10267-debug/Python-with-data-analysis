## Adding elements to a list

```python
countries = ['USA', 'USA', 'Canada', 'Mexico', 'Brazil', 'UK', 'France', 'Germany']
```

```python
countries.append('Italy')
print(countries)
```

```python
countries.insert(2, 'Argentina')
print(countries)
```

```python
countries_2 = ['Spain', 'Portugal']
countries = countries + countries_2
print(countries)
```

```python
nested_list = [countries, countries_2]
print(nested_list)
```
