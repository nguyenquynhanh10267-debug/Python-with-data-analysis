# If Statement

if <condition>:
    <code>
elif <condition>:
    <code>
    ....
else:
    <code>

```python
age = 15
if age>=18 :
    print("You're an adult!")
elif age >= 13:
    print("you're a teenager")
else:
    print("You're a kid")
```

## For loop

for <variable> in <iterable>:
    <code>

```python
countries
```

```python
for i, country in enumerate(countries):
    print(i)
    print(country)
```

```python
my_data = {'name':'Frank', 'age':28}
my_data
```

```python
for key, value in my_data.items():
    print (f"{key}:{value}")
```
